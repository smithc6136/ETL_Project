```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
twenty_eighteen_file = "csv_files/2018.csv"
twenty_eighteen_df = pd.read_csv(twenty_eighteen_file)
twenty_eighteen_df.head()
```

```python
twenty_nineteen_file = "csv_files/2019.csv"
twenty_nineteen_df = pd.read_csv(twenty_nineteen_file)
twenty_nineteen_df.head()
```

```python
# Create a filtered dataframe from specific columns
twentyeighteen_cols = ["Overall rank", "Country or region", "Score", "GDP per capita", "Healthy life expectancy"]
twentyeighteen_transformed= twenty_eighteen_df[twentyeighteen_cols].copy()

# Rename the column headers
twentyeighteen_transformed = twentyeighteen_transformed.rename(columns={"Overall rank": "rank",
                                                          "Country or region": "country_or_region",
                                                          "Score": "score",
                                                          "GDP per capita": "gdp_per_capita",
                                                          "Healthy life expectancy": "healthy_life_expectancy"})

# Clean the data by dropping duplicates and setting the index
twentyeighteen_transformed.drop_duplicates("rank", inplace=True)
twentyeighteen_transformed.set_index("rank", inplace=True)

twentyeighteen_transformed.head()
```

```python
# Create a filtered dataframe from specific columns
twentynineteen_cols = ["Overall rank", "Country or region", "Score", "GDP per capita", "Healthy life expectancy"]
twentynineteen_transformed= twenty_nineteen_df[twentynineteen_cols].copy()

# Rename the column headers
twentynineteen_transformed = twentynineteen_transformed.rename(columns={"Overall rank": "rank",
                                                          "Country or region": "country_or_region",
                                                          "Score": "score",
                                                          "GDP per capita": "gdp_per_capita",
                                                          "Healthy life expectancy": "healthy_life_expectancy"})

# Clean the data by dropping duplicates and setting the index
twentynineteen_transformed.drop_duplicates("rank", inplace=True)
twentynineteen_transformed.set_index("rank", inplace=True)

twentynineteen_transformed.head()
```

```python
connection_string = "catherine:etl_project@localhost:5432/etl_project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
twentyeighteen_transformed.to_sql(name='twentyeighteen', con=engine, if_exists='append', index=True)
```
